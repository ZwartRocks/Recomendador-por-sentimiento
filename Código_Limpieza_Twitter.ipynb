- **Carga de datos (pandas) y exportaci√≥n a fichero**.
- Concatena todos los dataframes en uno solo.
- Limpieza del texto.
- Post-proceso del dataset: **selecci√≥n/orden de columnas finales**, **renombrado de campos**.

```python
import pandas as pd
```

```python
archivo_name = [
    "T-Barcelona_tweets_20250810",
    "T-Gran_Canaria_tweets_20250811",
    "T-Madrid_tweets_20250810",
    "T-Malaga_tweets_20250810",
    "T-Palma_de_Mallorca_tweets_20250815",
    "T-Seville_tweets_20250811",
    "T-Tenerife_tweets_20250810",
    "T-Valencia_tweets_20250811"
]

# Lista para guardar los DataFrames temporales
df_list = []

for i in archivo_name:
    df_temp = pd.read_json(f"{i}.json")  # usar f-string para leer el archivo correcto
    df_list.append(df_temp)

# Concatenar todos los DataFrames en uno solo
df_final = pd.concat(df_list, ignore_index=True)


#== AQUI INCIA LA LIMPIEZA DE DATOS=============================================================================
#== AQUI INCIA LA LIMPIEZA DE DATOS=============================================================================
#== AQUI INCIA LA LIMPIEZA DE DATOS=============================================================================

# Extraer el texto del diccionario en la columna 'tweet'
df_final['Texto'] = df_final['tweet'].apply(lambda x: list(x.values())[0] if isinstance(x, dict) else x)

# Renombrar la columna 'keyword' a 'Experiencia'
df_final = df_final.rename(columns={'keyword': 'Experiencia'})

# Extraer solo la fecha (YYYY-MM-DD) de la columna 'fecha'
df_final['fecha'] = df_final['fecha'].str[:10]

# A√±adir nuevas columnas
df_final['Fuente'] = 'Twitter'
#df_final['Sentimiento'] = ''

# Reordenar columnas
df_final = df_final[['Texto', 'ciudad', 'Experiencia', 'fecha', 'Fuente']]

# Generar CSV
df_final["Texto"] = df_final["Texto"].str.replace("\n", " ", regex=True)
df_final.to_csv("Twitter.csv", index=False, sep="|", encoding="utf-8-sig")

# Mostrar el DataFrame final
df_final
```
