```python
import pandas as pd
```

```python
archivo_name = [
    "Barcelona_facebook_posts_20250811",
    "Gran_Canaria_facebook_posts_20250812",
    "Madrid_facebook_posts_20250811",
    "Malaga_facebook_posts_20250811",
    "Palma_de_Mallorca_facebook_posts_20250813",
    "Sevilla_facebook_posts_20250812",
    "Tenerife_facebook_posts_20250810",
    "Valencia_facebook_posts_20250813"
]

# Lista para guardar los DataFrames temporales
df_list = []

for i in archivo_name:
    df_temp = pd.read_json(f"{i}.json")  # usar f-string para leer el archivo correcto
    df_list.append(df_temp)

# Concatenar todos los DataFrames en uno solo
df_final = pd.concat(df_list, ignore_index=True)

# Extraer el texto del diccionario en la columna 'tweet'
df_final['texto'] = df_final['texto'].apply(lambda x: list(x.values())[0] if isinstance(x, dict) else x)

# Renombrar la columna 'keyword' a 'Experiencia'
df_final = df_final.rename(columns={'keyword': 'Experiencia'})

# Extraer solo la fecha (YYYY-MM-DD) de la columna 'fecha'
df_final['fecha'] = "08/2025"

# A침adir nuevas columnas
df_final['Fuente'] = 'Facebook'
#df_final['Sentimiento'] = ''

# Reordenar columnas
df_final = df_final[['texto', 'ciudad', 'Experiencia', 'fecha', 'Fuente']]

# Mostrar el DataFrame final
df_final.head(20)
```
