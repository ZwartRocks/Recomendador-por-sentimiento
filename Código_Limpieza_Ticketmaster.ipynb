### TRANSFORMACI√ìN Y LIMPIEZA

```python
import re
```

```python

# 1) Cargar archivo original
df = pd.read_csv("data_ticketmaster.csv")

# 2) Funci√≥n para limpiar texto
def clean_text(s):
    if pd.isna(s):
        return ""
    s = str(s)
    # Mantener letras, n√∫meros, puntuaci√≥n b√°sica y acentos
    s = re.sub(r"[^\w\s.,;:!?()\-\'\"√°√©√≠√≥√∫√º√±√Å√â√ç√ì√ö√ú√ë/@&]", " ", s)
    s = re.sub(r"\s+", " ", s).strip()  # quitar espacios m√∫ltiples
    return s

# 3) Definir clasificaci√≥n de experiencia
def classify_experience(row):
    tipo = str(row.get("tipo_evento", "") or "").lower()
    genero = str(row.get("genero", "") or "").lower()
    subgenero = str(row.get("subgenero", "") or "").lower()
    nombre = str(row.get("nombre", "") or "").lower()

    if "sport" in tipo or "deporte" in tipo:
        return "Deporte"
    if "music" in tipo or "m√∫sica" in tipo or "musica" in tipo:
        return "Vida nocturna"
    if "arts" in tipo or "theatre" in tipo or "teatro" in tipo or "cultura" in tipo:
        return "Cultura"
    if "film" in tipo or "cine" in tipo:
        return "Cultura"
    if "family" in tipo or "kids" in tipo or "familiar" in tipo:
        return "Ocio"

    # Fallbacks con palabras clave
    texto_busqueda = genero + " " + subgenero + " " + nombre
    if any(k in texto_busqueda for k in ["f√∫tbol", "futbol", "basket", "baloncesto", "tenis", "carrera", "marat√≥n", "maraton"]):
        return "Deporte"
    if any(k in texto_busqueda for k in ["concierto", "dj", "festival", "rock", "pop", "reggaeton", "rap", "techno", "electr√≥nica", "electronica", "club"]):
        return "Vida nocturna"
    if any(k in texto_busqueda for k in ["exposici√≥n", "exposicion", "museo", "galer√≠a", "galeria"]):
        return "Museo"
    if any(k in texto_busqueda for k in ["teatro", "√≥pera", "opera", "ballet", "danza", "comedia", "circo"]):
        return "Cultura"
    if any(k in texto_busqueda for k in ["tour", "visita", "ruta", "excursi√≥n", "excursion"]):
        return "Tour"
    if any(k in texto_busqueda for k in ["restaurante", "gastron", "tapas", "food", "comida", "vino", "cata"]):
        return "Comida"
    if any(k in texto_busqueda for k in ["playa", "beach", "chiringuito"]):
        return "Playa"
    if any(k in texto_busqueda for k in ["hotel", "resort", "spa"]):
        return "Hotel"

    return "Desconocido"

# 4) Crear nuevo DataFrame con columnas finales
data_ticketmaster = pd.DataFrame({
    "texto": df["nombre"].map(clean_text),
    "ciudad": df["ciudad_buscada"].fillna("").astype(str).str.strip(),
    "experiencia": df.apply(classify_experience, axis=1),
    "fecha": pd.to_datetime(df["fecha"], errors="coerce").dt.date,
    "fuente": "Ticketmaster"
})

# üîÑ Normalizar nombres de ciudades
reemplazos_ciudad = {
    "Palma de Mallorca": "Mallorca",
    "Santa Cruz de Tenerife": "Tenerife",
    "Las Palmas de Gran Canaria": "Gran Canaria"
}
data_ticketmaster["ciudad"] = data_ticketmaster["ciudad"].replace(reemplazos_ciudad)

# 5) Guardar resultado
data_ticketmaster.to_csv("ticketmaster_final.csv", index=False, encoding="utf-8")

print("‚úÖ Archivo guardado como ticketmaster_final.csv")
print(data_ticketmaster.head())
print("\nüìä Conteo por ciudad normalizada:")
print(data_ticketmaster["ciudad"].value_counts())
```
